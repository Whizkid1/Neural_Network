mkdir sub1/sub2/sub3 : sub1안에 sub2안에 sub3 디렉토리 만들기(sub2가 이미 만들어져 있을 때) \
mkdir -p sub1/sub2/sub3 : 위와 동일(단, sub2도 함께 새로 만든다) \
